```python
import batsim
import galsim
import numpy as np
```

```python
sb_obj = galsim.Sersic(n=4, half_light_radius=0.5)

print(sb_obj._sbp)

batsim._gsinterface.getFluxVec(scale=0.2, gsobj=sb_obj._sbp, xy_coords=np.array([[0.1, 0.1], [0.2, 0.2]]))
```

```python
trans_obj = galsim.Sersic(n=4, half_light_radius=0.5).shear(g1=0.1, g2=0.2).shift(0.5, 0.5)

print(trans_obj._sbp)

batsim._gsinterface.getFluxVec(scale=0.2, gsobj=sb_obj._sbp, xy_coords=np.array([[0.1, 0.1], [0.2, 0.2]]))
```
